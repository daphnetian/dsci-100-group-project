**Q2.** What do the variable names mean? Use the manual documentation to find out!

```r
# ?read_csv
read<-read_csv("processed.cleveland.data")
names(read)<-c("age", "sex", "chest_pain_type", "resting_bp", "chol", "fasting_blood_sugar", "resting_electrocardiographic_results", "maximum_heart_rate_achieved", "exercise_induced_angina", "oldpeak", "slope", "ca", "thal", "num")
readM<- read  %>%
filter(sex == 1) %>%
mutate(sex = "Male")

readF<- read %>%
filter(sex == 0) %>%
mutate(sex = "Female")

rbind(readM, readF)
# readM
# readF


my_plot<-read %>%
       ggplot(aes(x = age, y = chol, color = sex)) +
       geom_point()
my_plot

# read
```
